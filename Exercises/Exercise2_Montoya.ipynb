### Creating a real weekly wage variable in November 2020 dollars.

```stata
generate real_weekly_wage = ${cpindexNov20}*earnweek/cpi // real weekly wage in Nov2020 dollars
//scalar lcut=$mwage20 * 40/2 // half weekly earnings from 40 hours/week at the 2020 minimum wage
//drop if real_weekly_wage<lcut

summarize real_weekly_wage
```

## D. Create and Plot a Data Set of Statistics.

### Collapse the data to a set of annual statistics: the mean, the 10th, 50th, and 90th percentiles of the real weekly wage

```stata
collapse (mean) mean=real_weekly_wage (p10) p10=real_weekly_wage (p50) p50=real_weekly_wage (p90) p90=real_weekly_wage, by(year)
describe
```

### Plot the four series from 1982 to 2020.

```stata
generate y=year-1
line p10 y || line p50 y || line p90 y || line mean y, ///
legend(on label(1 "10th") label(2 "50th") label(3 "90th") label (4 "Mean") rows(1)) name(Fig0, replace) /// 
title("Real weekly wages 1982 - 2020")
graph export "..\Analysis\Results\real_weekly_wages.pdf", as(pdf) replace
```

## D. Create and Plot a Data Set of Statistics.

### Collapse the data to a set of annual statistics: the mean, the 10th, 50th, and 90th percentiles of the real weekly wage

```stata
collapse (mean) mean=real_weekly_wage (p10) p10=real_weekly_wage (p50) p50=real_weekly_wage (p90) p90=real_weekly_wage, by(year)
describe
```

### Plot the four series from 1982 to 2020.

```stata
generate y=year-1
line p10 y || line p50 y || line p90 y || line mean y, ///
legend(on label(1 "10th") label(2 "50th") label(3 "90th") label (4 "Mean") rows(1)) name(Fig0, replace) /// 
title("Real weekly wages 1982 - 2020")
graph export "..\Analysis\Results\real_weekly_wages.pdf", as(pdf) replace
```

Based on the grapgh above, we can observe that weekly wages have stagnated for the 10th and 50th percentile, whereas for the 90th percentile, they have increased.

### Log difference between the 90th and 10th percentiles of the real weekly wage.

```stata
generate logdiff = log(p90) - log(p10)
line logdiff y, legend( on label (1 "Log difference between 90th & 10th percentile") rows(1)) name(Fig1, replace) ///
title("Log difference of real weekly wages 1982 - 2020")
graph export "..\Analysis\Results\log_real_weekly_wages_.pdf", as(pdf) replace
```

Based on the grapgh above, we can observe that log difference of the real weekly wage between the 90th and 10th percentile have increased as the years have gone by. 

## Percentiles

```stata
foreach y of numlist 1982 1983/2020 {
      pctile y`y' = real_weekly_wage if year==`y', nq(100)
    }
keep if y1982~=.
keep y1982-y2020
generate percentile=_n
order percentile, first
describe
summarize
```

```stata
reshape long y, i(percentile) j(year)
rename y real_weekly_wage
sort year percentile
order year percentile real_weekly_wage
format real_weekly_wage %7.2fc
```
