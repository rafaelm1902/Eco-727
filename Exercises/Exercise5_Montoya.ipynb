## B. Create Dummy and Dummy-Interaction Variables.

### Dummy variables

```stata
/* Variable for the post-hike period 1990*/
generate post_hike90 = y>=1990
tab post_hike90

/* Variable for the post-hike periods after 1991*/
generate post_hike91 = y>=1991
tab post_hike91

/* Variable for minimum wage year 1990*/
generate mw_year90 = y==1990
tab mw_year90
```

### Dummy interaction variables

```stata
/* Interaction variable for pos90 period and adult wages */
generate adult_int90 = post_hike90 * med_wage_oth

/* Interaction variable for pos90 period and teen wages */
generate teen_int90 = post_hike90 * med_wage_teen

/* Third treatment variable for pre-event proportion of teens in the treatment zone between the 1989 & 1991 minimum wages*/
generate third_treatment = treatment90 + treatment91
generate secint90 = post_hike90 * third_treatment
```

## C. Regress employed on Post-Event Dummies

### In a first regression, estimate a single effect in the post-event period

```stata
regress employed post_hike90 i.age i.state [pw=pweight], vce(cluster state)
```

We observe that the post variable has a negative coefficient, signifying a decrease in teen employment after the 1990 hike. This not enough to conclude that the hike caused lower employment. 

### In a secondregression, estimate separate effects for (a) 1990 and (b) subsequent years

```stata
regress employed mw_year90 post_hike91 i.age i.state [pw=pweight], vce(cluster state)
```

We add a minimun wage 1990 variable, and a variable for minimum wage year 1991 and after, the coeffiecient for each varaible is negative. Signifying that teen employment is lower after each hike. 

### Test the hypothesis of equal employment effects in 1990 and later

```stata
test mw_year90 = post_hike91
```

We test the null hypothesis that employment effects in 1990 are equal to those in 1991 and after. We fail to reject the null at the 5% level, however at the 10% level  we reject the null can find that employment effects in 1990 and subsequent year are not equal.  

### Add a linear time trend in a third regression

```stata
/* I was not sure if it was necesarry to create a time variable or if the use of the varaible y was enough for a time trend*/
/* Nonetheless, I created a time varaible*/

/* We create a time varaible equal to months since april 1985 */
generate time = ym(year, month) - 303
summarize time
```

```stata
/*Sorry for two regressions instead of one, as I was not sure which one to use. */

regress employed mw_year90 post_hike91 i.age i.state time [pw=pweight], vce(cluster state) // regression with time for linear trend

regress employed mw_year90 post_hike91 i.age i.state y [pw=pweight], vce(cluster state) // regression with y for linear trend
```

### Does including the trend increase or decrease the estimated effects on teenemployment? Are the estimates in this section DD estimates or what?

For both of the regressions, adding a linear time trend increases the estimated effects on teen employment. We also notice that that the trend variable is positive for both regressions, signifying that eployment is trending up. The estimates in this section are not DD, since we do not include an interaction varaible between the treatment and the post variable. Nonetheless, this estimates helps us understand the effects of the different variables used on employment.

## D. Regress employed on Measures of Treatment Intensity

### (1) A teen-employment regression with age, year, and state effects

```stata
regress employed i.age i.year i.state [pw=pweight], vce(cluster state)
```

We see that older teens have higher employment and observe year effects where all coefficients are negative after 1990.

### Interpret the estimated interaction coefficient. Is the evidence consistent with the idea that minimum-wage hikes reduce teen employment? Are these DD estimates or what?

Based on the estimates, the proportion of teens that were on the pre event treatment zone after 1990 have higher employment. On avarege, employment would increase by .25 points. However, if we look at the other interaction between median wages and teens post90 show that there is a reduction in employment which is more consisten with the literature. These estimates are DD since we include pre-event proportion of teens in secint90 as well is an indicator (teen_int90) of whether a teen is from before or after the wage hike.
