## Estimate the log-Wage Regression by OLS 

### Regress the weekly wage variable (in logs) on race.

```stata
eststo:reg log_rwage black [pw=earnwt]
```

Without including anyother explanatory variables, we can see that on average black individuals earn 21.58% less than white individuals

```stata
eststo: reg log_rwage black grade experience experience_squared female i.region in_metro married single [pw=earnwt]
```

we add industry and occupation

```stata
eststo: reg log_rwage black grade experience experience_squared female i.region in_metro married single /// 
i.industry i.occupation [pw=earnwt]
```

We add union

```stata
eststo: reg log_rwage black grade experience experience_squared female i.region in_metro married single union /// 
i.industry i.occupation [pw=earnwt]
```

We add cohort effects, birthyear

```stata
eststo: reg log_rwage black grade experience experience_squared female i.region in_metro married single union /// 
i.industry i.occupation i.birthyear [pw=earnwt]
```

We add highschool bachelors advancedegree

```stata
eststo: reg log_rwage black grade experience experience_squared female i.region in_metro married single union /// 
i.industry i.occupation i.birthyear highschool bachelors advancedegree [pw=earnwt]
```

## We create tables 

```stata
esttab est1 est2 est3 est4 est5 est6, replace nodepvars nostar nomtitles compress /// 
b(3) se(3) stats(N r2, fmt(%3.0f %4.3f) labels("N" "R-sq")) /// 
coeflabels( _cons "Constant" black "Black"  grade "Highest Grade Completed" experience "Experience" /// 
experience_squared "Experience Squared" female "Female" in_metro "Metro" married "Married" ///
single "Single" union "Union" highschool "H.S Diploma Only" bachelors "Bachelor's Only" /// 
advancedegree "Master's/Professional/Doctorate Only") order(_cons) /// 
indicate("Region Effects = *.region" "Idustry Effects= *.industry"  /// 
"Occupation Effects = *.occupation" "Birth Cohorts = *.birthyear") alignment(r) ///
addnotes( "The independent variable is the log of weekly wages in all regressions" ///
"SOURCE: Current Population Survey and author's calculations")
```

```stata
esttab est1 est2 est3 est4 est5 est6 using Results/table_2.tex, replace nodepvars nostar nomtitles booktabs compress /// 
b(3) se(3) stats(N r2, fmt(%3.0f %4.3f) labels("N" "R-sq")) /// 
coeflabels( _cons "Constant" black "Black"  grade "Highest Grade Completed" experience "Experience" /// 
experience_squared "Experience Squared" female "Female" in_metro "Metro" married "Married" ///
single "Single" union "Union" highschool "H.S Diploma Only" bachelors "Bachelor's Only" /// 
advancedegree "Master's/Professional/Doctorate Only") order(_cons) /// 
indicate("Region Effects = *.region" "Idustry Effects= *.industry"  /// 
"Occupation Effects = *.occupation" "Birth Cohorts = *.birthyear") alignment(r) ///
addnotes("The independent variable is the log of weekly wages in all regressions" ///
"SOURCE: Current Population Survey and author's calculations")
```

## Oaxaca-Blinder Decomposition

```stata
svyset [pw=earnwt] 
```

```stata
eststo: oaxaca log_rwage grade highschool bachelors advancedegree experience experience_squared female region in_metro married ///
single union industry occupation birthyear, by(black) pooled svy
```

```stata
esttab est7, replace nodepvars nostar nonumbers not noobs  /// 
keep(group_1 group_2 difference explained unexplained grade industry occupation) ///
b(3) coeflabels(  group_1 "White weekly wage(log)" group_2 "Black weekly wage(log)" difference "Difference" ///
explained "Explained by observed differences in controls" unexplained "Unexplained by observed differences in controls" ///
grade "Highest Grade Completed" industry "Industry Effects" occupation "Ocuppation Effects") ///
mtitles(" All Controls")  alignment(r) ///
addnotes("Percent of wage difference explained equal to 43 percent" ///
"SOURCE: Current Population Survey and author's calculations")
```
