# 2. Adressing missing observations & recoding for consistency

We drop some of the variables we won't be using and recode for consistency as well as missing observations. 

### Addressing missing observations


```stata
replace region=. if region==97
replace sex=. if sex==9
replace race=. if race==999
replace marst=. if marst==9 // NIU
replace metro=. if inlist(metro, 0, 9) // Not identifiable, Missing/Unknown
replace classwkr=. if inlist(classwkr, 00, 99) // NIU, Missing/Unknown
replace empstat=. if empstat==00 
replace labforce=. if labforce==0
replace educ=. if inlist(educ, 000, 001, 999) // NIU or no schooling, NIU or blank, Missing/Unknown
replace union=. if union==0 
replace occ1990 =. if inlist(occ1990, 991, 999) // Unemployed, N/A and unknown 
replace ind1990=. if inlist(ind1990, 000, 997,998) // NIU, Unknown
replace earnweek =. if earnweek == 9999.99
```

### Recoding Occupation

We procceed by recoding occupation. We generate occupation to replace occ1990 and we code for a broad set of occupations.

```stata
generate occupation=.

*** 1982-2020
replace occupation=1 if inrange(occ1990, 000, 037) & inrange(year, 1982, $eyear) // Executive, Administrative, and Managerial Occupations 
replace occupation=2 if inrange(occ1990, 038, 200) & inrange(year, 1982, $eyear) // Professional Specialty Occupations
replace occupation=3 if inrange(occ1990, 201, 235) & inrange(year, 1982, $eyear) // Technicians and Related Support Occupations
replace occupation=4 if inrange(occ1990, 243, 290) & inrange(year, 1982, $eyear) // Sales Occupations
replace occupation=5 if inrange(occ1990, 303, 400) & inrange(year, 1982, $eyear) // Administrative Support Occupations 
replace occupation=6 if inrange(occ1990, 401, 470) & inrange(year, 1982, $eyear) // Service Occupations
replace occupation=7 if inrange(occ1990, 471, 500) & inrange(year, 1982, $eyear) // Farming, Forestry, and Fishing  
replace occupation=8 if inrange(occ1990, 501, 700) & inrange(year, 1982, $eyear) // Precision Production, Craft, and Repair
replace occupation=9 if inrange(occ1990, 701, 900) & inrange(year, 1982, $eyear) // Operators, Fabricators, and Laborers
tab year occ1990 if occ1990~=. & occupation==. & inrange(year, 1982, $eyear)
```

### Recoding Industry

We procceed by recoding industry. We generate industry to replace ind1990 and we code for 13 industries.

```stata
generate industry=.

*** 1982-2020
replace industry=1 if inrange(ind1990, 010, 032) & inrange(year, 1982, $eyear) // Agriculture, Forestry, and Fisheries
replace industry=2 if inrange(ind1990, 040, 050) & inrange(year, 1982, $eyear) // Mining
replace industry=3 if ind1990 == 060             & inrange(year, 1982, $eyear) // Construction
replace industry=4 if inrange(ind1990, 100, 392) & inrange(year, 1982, $eyear) // Manufacturing 
replace industry=5 if inrange(ind1990, 400, 472) & inrange(year, 1982, $eyear) // Transportation, Communications, and Utilities
replace industry=6 if inrange(ind1990, 500, 571) & inrange(year, 1982, $eyear) // Wholesale trade
replace industry=7 if inrange(ind1990, 580, 691) & inrange(year, 1982, $eyear) // Retail trade
replace industry=8 if inrange(ind1990, 700, 712) & inrange(year, 1982, $eyear) // Finance, Insurance, and Real Estate
replace industry=9 if inrange(ind1990, 721, 760) & inrange(year, 1982, $eyear) // Business and Repair Services
replace industry=10 if inrange(ind1990, 761, 791) & inrange(year, 1982, $eyear) // Personal Services
replace industry=11 if inrange(ind1990, 800, 810) & inrange(year, 1982, $eyear) // Entertainment and Recreation Services
replace industry=12 if inrange(ind1990, 812, 893) & inrange(year, 1982, $eyear) // Professional and Related Services 
replace industry=13 if inrange(ind1990, 900, 932) & inrange(year, 1982, $eyear) // Public Administration
tab year ind1990 if ind1990~=. & industry==. & inrange(year, 1982, $eyear)
```

### Recoding Education

We first recode the education variable educ to create the variable grade, which measures highest grade completed. We use Jaeger (1997) to transform degrees into grades.

```stata
generate grade=.
*** 1982-1991
  replace grade=0  if educ==2 & year<=1991 // none or preschool
  replace grade=1  if educ==11 & year<=1991 // grade 1
  replace grade=2  if educ==12 & year<=1991 // grade 2
  replace grade=3  if educ==13 & year<=1991 // grade 3
  replace grade=4  if educ==14 & year<=1991 // grade 4
  replace grade=5  if educ==21 & year<=1991 // grade 5
  replace grade=6  if educ==22 & year<=1991 // grade 6
  replace grade=7  if educ==31 & year<=1991 // grade 7
  replace grade=8  if educ==32 & year<=1991 // grade 8 
  replace grade=9  if educ==40 & year<=1991 // grade 9
  replace grade=10 if educ==50 & year<=1991 // grade 10
  replace grade=11 if educ==60 & year<=1991 // grade 11
  replace grade=12 if (educ==72 | educ==73) & year<=1991 // 12th grade, diploma unclear, 
  replace grade=13 if educ==80 & year<=1991 // 1 year of college
  replace grade=14 if educ==90 & year<=1991 // 2 years of college
  replace grade=15 if educ==100 & year<=1991 // 3 years of college
  replace grade=16 if educ==110 & year<=1991 // 4 years of college
  replace grade=17 if educ==121 & year<=1991 // 5 years of college
  replace grade=18 if educ==122 & year<=1991 // 6 years of college or more
  
  *** 1992-2109 (Jaeger 1997)
  replace grade=0   if educ==2  & year>=1992 // none or preschool 
  replace grade=2.5 if educ==10 & year>=1992 // grades 1, 2, 3, or 4
  replace grade=5.5 if educ==20 & year>=1992 // grades 5 or 6
  replace grade=7.5 if educ==30 & year>=1992 // grades 7 or 8
  replace grade=9   if educ==40 & year>=1992 // grade 9
  replace grade=10  if educ==50 & year>=1992 // grade 10
  replace grade=11  if educ==60 & year>=1992 // grade 11
  replace grade=12  if (educ==71 | educ==73) & year>=1992 // 71 is `12th grade, no diploma' 73 is 'HS diploma or equivalent'
  replace grade=13  if educ==81 & year>=1992 // some college, no degree
  replace grade=14  if (educ==91 | educ==92) & year>=1992 // 91 is 'vocational associate's degree' 92 is 'academic a.d' 
  replace grade=16  if educ==111 & year>=1992 // bachelor's degree
  replace grade=18  if inlist(educ, 123, 124, 125) & year>=1992 // master's, professional school, doctorate degrees
  order grade, before(educ)

** We label grade
  label value grade GRADE
  label variable grade  "highest grade completed"
```

### Generate a variable to identify top-coded values of earnweek

```stata
/*Identifying top coded values */
gen topcode_earnweek = .
replace topcode_earnweek = 0 if earnweek != . & inrange(year, 1982,$eyear)
replace topcode_earnweek = 1 if earnweek == 999 & inrange(year, 1982, 1988)
replace topcode_earnweek = 1 if earnweek== 1923 & inrange(year, 1989, 1997)
replace topcode_earnweek = 1 if earnweek == 2884.61 & inrange(year, 1998,$eyear)
label variable topcode_earnweek "top Coded Values of weekly earnings" // To rename variable labels
label define TOPCODE_EARNWEEK 0 "0-Not Top Coded" 1 "1-Top Coded" // To rename labels, so it's not just 1 and 0 
label value topcode_earnweek TOPCODE
tab year topcode_earnweek, missing
```

### Generating real weekly wages and the log of real weekly wages

```stata
/*Generating the real weekly wage  */
summarize cpi if inlist(year, 2020) & inlist(month, 12), meanonly
scalar cpidec20=r(mean)
scalar list cpidec20
generate real_weekly_wage = cpidec20*earnweek/cpi // real weekly wage in Dec2020 dollars
scalar lcut=$mwage20 * 40/2 // half weekly earnings from 40 hours/week at the 2020 minimum wage
drop if real_weekly_wage<lcut
generate log_rwage = log(real_weekly_wage)

label value real_weekly_wage RWWAGE
label variable real_weekly_wage "real weekly wage, Dec2020 dollars"
label value log_rwage LOGWAGE
label variable log_rwage " log of real weekly wage"

summarize real_weekly_wage log_rwage
```


### Generating variables
